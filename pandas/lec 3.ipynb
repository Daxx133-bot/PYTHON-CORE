```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_excel("IPL.xls")
df.head()
```

```python
df.info()
```

```python
one = df[df['Winner'] == "Match tied"]
```

```python
two = df[df['Balls_Remaining'] <= 3]
```

```python
three = df[
    (df['Winning_Team'] == "Chasing") &
    (df['Bat_Second_Wickets_Lost'] >= 9)
]
```

```python
four = df[
    (df['Winning_Team'] == "FirstBatting") &
    (df['Winning_Margin'] <= 5)
]
```

```python
close_matches = pd.concat([one, two, three, four])
close_matches
```

```python
close_matches[close_matches["Year"] == 2008]
```

```python
close_matches.groupby('Year').size()
```

```python
teams = [
    'Chennai',
    'Hyderabad',
    'Mumbai',
    'Kolkata'
]

for team in teams:

    A = len(df[df['Winner'] == team])

    B = len(
        df[
            (df['Winner'] == team) &
            (df['Winning_Team'] == 'FirstBatting')
        ]
    )

    percent = (B / A) * 100

    print(team, round(percent,2))
```
